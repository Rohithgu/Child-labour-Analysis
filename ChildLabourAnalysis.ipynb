```python
import seaborn as sns
plt.figure()
sns.lineplot(data=child, x="Year", y="Child_Labor_Rate_Percent")
plt.title("Child Labor Rate Trend Over Years")
plt.show()
```

```python
child = child.rename(columns={'Age_Group': 'Age_groups'})
```

```python
child
```

```python
plt.bar(data=child,
        x=child['Year'],
        height=child['Child_Labor_Rate_Percent'])
plt.title("Child Labour Over Years")
plt.show()
```

```python
plt.bar(data=child,
        x=child['Age_groups'],
        height=child['Child_Labor_Rate_Percent'])
plt.title("Child Labor Rate by Age Group")
plt.show()
```

```python
plt.figure()
sns.boxplot(
    data=child,
    x="Region",
    y="Child_Labor_Rate_Percent",
    showmeans=False
)
plt.title("Child Labor Rate by Region")
plt.show()
```

```python
child.loc[(child['Age_groups']=='7-14 years')]
```

```python
child.loc[(child['Age_groups']=='5-17 years')]
```

```python

plt.figure(figsize=(12,7))
sns.violinplot(
    data=child,
    x="Region",
    y="Child_Labor_Rate_Percent",
    inner=None,
    cut=0
)

sns.swarmplot(
    data=child,
    x="Region",
    y="Child_Labor_Rate_Percent",

)

plt.title("Region-wise Child Labor Rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(
    child["Estimated_Children_Millions"],
    child["Child_Labor_Rate_Percent"],
    s=60,
    alpha=0.6
)

plt.xlabel("Estimated Children (Millions)")
plt.ylabel("Child Labor Rate (%)")
plt.title("Estimated Children vs Child Labor Rate (Circle Scatter)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(
    data=child,
    x="Year",
    y="Child_Labor_Rate_Percent",
    cut=0
)

plt.xticks(rotation=45)
plt.title("Year-wise Distribution of Child Labor Rate")
plt.show()
```

```python
labour=child.groupby(by=['Year']).agg(
    labour_rate=("Child_Labor_Rate_Percent", "mean")
)
```

```python
labour
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=labour,
    x="Year",
    y="labour_rate"
)

plt.xticks(rotation=45)
plt.title("Child Labour Rate Year-Wise")
plt.xlabel("Year")
plt.ylabel("Labour Rate (%)")
plt.show()
```
